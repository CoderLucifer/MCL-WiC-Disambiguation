```python
import os
import json
import pandas as pd

def agg(data):
    lemma, text1, text2, word1, word2, lang, pos = [], [], [], [], [], [], []
    ids = []
    for d in data:
        ids.append(d['id'])
        text1.append(d['sentence1'])
        text2.append(d['sentence2'])
        try:
            word1.append(d['sentence1'][int(d['start1']):int(d['end1'])])
            word2.append(d['sentence2'][int(d['start2']):int(d['end2'])])
        except:
            print(d['id'], d['ranges1'], d['ranges2'])
            ranges1 = d['ranges1'].split(',')
            ranges2 = d['ranges2'].split(',')
            
            span1, span2 = [], []
            
            for x in ranges1:
                s, e  = x.split('-')
                span1.append(d['sentence1'][int(s):int(e)])
            for x in ranges2:
                s, e  = x.split('-')
                span2.append(d['sentence2'][int(s):int(e)])
            
            span1 = ' '.join(span1)
            span2 = ' '.join(span2)
            
            word1.append(span1)
            word2.append(span2)
            
        pos.append(d['pos'])
        lemma.append(d['lemma'])
        lang.append(d['id'].split('.')[1])
    
    df = pd.DataFrame({'id': ids,
                        'lang':lang,
                        'word1': word1,
                        'sentence1': text1,
                        'word2': word2,
                        'sentence2': text2,
                        'lemma': lemma,
                        'pos': pos,
                        })
    
    df.loc[:, 'target'] = 'N'
    return df


files = os.listdir('test')

dfs = []

for f in files:
    if 'gold' not in f and 'trial' not in f:
        data = json.load(open(os.path.join('test', f)))
        print(f)
        dfs.append(agg(data))


df = pd.concat(dfs, axis=0)
print(df.shape)
df = df.drop_duplicates()
print(df.shape)
df.to_csv('./csvs/test_data.csv', index=False)
```
